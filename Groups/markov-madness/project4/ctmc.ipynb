```python
from sympy import zeros, ones
from collections import namedtuple
states = ("operational", "degrading", "unresponsive")
TraceItem = namedtuple("TraceItem", "component state")
components = (
    'Authentication Service', 'Availability Item Filter', 'Bid and Buy Service',
    'Buy Now Item Filter',
    'Category Item Filter', 'Comment Item Filter', 'Future Sales Item Filter', 'Inventory Service',
    'Item Management Service', 'Last Second Sales Item Filter', 'Past Sales Item Filter',
    'Persistence Service', 'Query Service', 'Recommendation Item Filter', 'Region Item Filter',
    'Reputation Service', 'Seller Reputation Item Filter', 'User Management Service', 'Supervisory Component')

with open('data/traces-30.txt', 'r') as file:
    traces = eval(file.read())

G = zeros(len(components) * len(states))

traces = [[components.index(x.component) * 3 + states.index(x.state) for x in trace] for trace in traces]

for trace in traces:
    for i, item in enumerate(trace):
        if i < len(trace) - 1:
            if item != trace[i+1]:
                G[item, trace[i+1]] += 1
sums = G * ones(G.shape[1], 1)
for i, x in enumerate(sums):
    G[i, i] = -x
assert (G * ones(G.shape[1], 1)).T == zeros(1, G.shape[1])
G
```

```python
C = ContinuousMarkovChain('C', state_space=[x for x in range(len(components) * len(states))], gen_mat=G)
ld = C.limiting_distribution().evalf()
ld.T
```
